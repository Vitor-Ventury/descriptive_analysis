# Problemática

Trabalharemos com os dados do sistema de nascidos vivos do DataSUS do ano de 2019 do estado de Rondônia, onde traremos dados descritivos a fim de obter informações relevantes

# Importar bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib notebook
```

# Leitura dos dados

```python
df = pd.read_csv('./sinasc_ro_2019.csv')

#Exibir colunas ocultas
pd.set_option('max_columns', None)
df.head(5)
```

# Obter informações descritivas das colunas de interesse

Os dados podem ser obtidos de forma individual para melhor visualização e análise ou de forma conjunta para otimizar o tempo

```python
fig, ax = plt.subplots(1,1)
sns.set_style(style='darkgrid')
sns.countplot(x=df['APGAR1'], data=df)
ax.set_ylabel('CONTAGEM')
ax.set_title('RELAÇÃO APGAR1')
ax.tick_params(axis='x', labelsize='small')
ax.tick_params(axis='y', labelsize='small')
```

```python
fig, ax = plt.subplots(1, 1)
sns.countplot(x=df['APGAR5'], data=df)
ax.set_ylabel('CONTAGEM')
ax.set_title('RELAÇÃO APGAR5')
ax.tick_params(axis='x', labelsize='small')
ax.tick_params(axis='y', labelsize='small')
```

```python
fig, ax = plt.subplots(1, 1)
sns.countplot(x=df['SEXO'], data=df)
ax.set_ylabel('CONTAGEM')
ax.set_title('RELAÇÃO SEXO')
ax.tick_params(axis='x', labelsize='small')
ax.tick_params(axis='y', labelsize='small')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10,8))
sns.countplot(x=df['GESTACAO'], data=df)
ax.set_title('RELAÇÃO GESTAÇÃO')
ax.set_ylabel('CONTAGEM')
ax.tick_params(axis='x', labelrotation=35, labelsize='small')
ax.tick_params(axis='y', labelsize='small')
```

```python
fig, ax = plt.subplots(1, 1)
sns.countplot(x=df['ESTCIVMAE'], data=df)
ax.set_title('RELAÇÃO ESTADO CIVIL')
ax.set_ylabel('CONTAGEM')
ax.tick_params(axis='x', labelsize='small')
ax.tick_params(axis='y', labelsize='small')
```

# Exibir APGAR5 médio de forma descritiva

```python
fig, ax = plt.subplots(1, 1)
sns.barplot(x='CONSULTAS',y='APGAR5', data=df)
ax.set_title('RELAÇÃO APGAR5 X CONSULTAS')
ax.tick_params(axis='x', labelsize='small')
ax.tick_params(axis='y', labelsize='small')
```

# Construir gráficos conjuntos com colunas de interesse

```python
fig, axes = plt.subplots(2, 2, figsize=(10,12))
sns.countplot(ax=axes[0,0], x=df['SEXO'], data=df)
sns.countplot(ax=axes[0,1], x=df['ESTCIVMAE'], data=df)
sns.countplot(ax=axes[1,0], x=df['GESTACAO'], data=df)
sns.countplot(ax=axes[1,1], x=df['CONSULTAS'], data=df)

# Implementar melhorias nos gráficos
axes[0,0].set_xlabel('')
axes[0,0].set_ylabel('')
axes[0,0].set_title('SEXO')
axes[0,0].tick_params(axis='x', labelsize='small')
axes[0,0].tick_params(axis='y', labelsize='small')

axes[0,1].set_xlabel('')
axes[0,1].set_ylabel('')
axes[0,1].set_title('ESTCIVMAE')
axes[0,1].tick_params(axis='x', labelsize='small')
axes[0,1].tick_params(axis='y', labelsize='small')

axes[1,0].set_xlabel('')
axes[1,0].set_ylabel('')
axes[1,0].set_title('GESTACAO')
axes[1,0].tick_params(axis='x', labelsize='small', rotation=90)
axes[1,0].tick_params(axis='y', labelsize='small')

axes[1,1].set_xlabel('')
axes[1,1].set_ylabel('')
axes[1,1].set_title('CONSULTAS')
axes[1,1].tick_params(axis='x', labelsize='small')
axes[1,1].tick_params(axis='y', labelsize='small')
```
